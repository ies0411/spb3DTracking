```python
# 0 1 Cyclist -1 -1 -10 720.5727208072406 303.690263264956 943.8370166622524 374.0 1.5672513246536255 0.5660192966461182 1.7846254110336304 6.077325344085693 -1.6074483394622803 -0.8267881870269775 6.458888530731201
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_path ='../sample/image/0000/000000.png'
image = cv2.imread(img_path)
```

```python
plt.subplots(1, 1, figsize=(15, 7))
plt.imshow(image)
```

```python
# 바운딩 박스를 그릴 좌표 (x1, y1), (x2, y2)

# 0 1 Cyclist -1 -1 -10 720.5727208072406 303.690263264956 943.8370166622524 374.0 1.5672513246536255 0.5660192966461182 1.7846254110336304 6.077325344085693 -1.6074483394622803 -0.8267881870269775 6.458888530731201 


x1, y1, x2, y2 = 720, 303, 943, 374.

# 바운딩 박스 색상 및 두께 설정
color = (0, 255, 0)  # 초록색 (BGR 형식)
thickness = 2  # 선 두께

# 이미지에 바운딩 박스 그리기
image_with_bbox = cv2.rectangle(image, (x1, y1), (x2, y2), color, thickness)
```

```python
plt.subplots(1, 1, figsize=(15, 7))
plt.imshow(image_with_bbox)
```
